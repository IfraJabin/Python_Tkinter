```python
from tkinter import *
```

```python
def exit():
    window.destroy()
```

```python
def clicked(number):
    x=enter_number.get()
    enter_number.delete(0,END)
    enter_number.insert(0,str(x)+str (number))
    
def clear():
    enter_number.delete(0,END)
    
def plus():
    num_1=enter_number.get()
    global first
    global math
    math='add'
    first=int(num_1)
    enter_number.delete(0,END)
    
def minus():
    num_1=enter_number.get()
    global first
    global math
    math='minus'
    first=int(num_1)
    enter_number.delete(0,END)
    
def multiply():
    num_1=enter_number.get()
    global first
    global math
    math='multiply'
    first=int(num_1)
    enter_number.delete(0,END)
    
def divide():
    num_1=enter_number.get()
    global first
    global math
    math='divide'
    first=int(num_1)
    enter_number.delete(0,END)


def equal():
        num_2=enter_number.get()
        enter_number.delete(0,END)
        if math=='add':
            enter_number.insert(0,first+int(num_2))
        elif math=='minus':
            enter_number.insert(0,first-int(num_2))
        elif math=='multiply':
            enter_number.insert(0,first*int(num_2))
        elif math=='divide':
            enter_number.insert(0,first/int(num_2))
        
```

```python
window=Tk()
window.geometry('335x465')
window.title('Calculator')


enter_number=Entry(window)
enter_number.place(x=0,y=25,width=335,height=100)


close=Button(window,text='X',command=exit)
close.place(x=315,y=0)

clear=Button(window,text='C',width=10,height=5,command=clear)
clear.place(x=0,y=380)
one=Button(window,text='1',width=10,height=5,command=lambda: clicked(1))
one.place(x=0,y=295)
four=Button(window,text='4',width=10,height=5,command=lambda: clicked(4))
four.place(x=0,y=210)
seven=Button(window,text='7',width=10,height=5,command=lambda: clicked(7))
seven.place(x=0,y=125)

zero=Button(window,text='0',width=10,height=5,command=lambda: clicked(0))
zero.place(x=85,y=380)
two=Button(window,text='2',width=10,height=5,command=lambda: clicked(2))
two.place(x=85,y=295)
five=Button(window,text='5',width=10,height=5,command=lambda: clicked(5))
five.place(x=85,y=210)
eight=Button(window,text='8',width=10,height=5,command=lambda: clicked(8))
eight.place(x=85,y=125)

equal=Button(window,text='=',width=10,height=5,command=equal)
equal.place(x=170,y=380)
three=Button(window,text='3',width=10,height=5,command=lambda: clicked(3))
three.place(x=170,y=295)
six=Button(window,text='6',width=10,height=5,command=lambda: clicked(6))
six.place(x=170,y=210)
nine=Button(window,text='9',width=10,height=5,command=lambda: clicked(9))
nine.place(x=170,y=125)

divide=Button(window,text='/',width=10,height=5,command=divide)
divide.place(x=255,y=380)
multiply=Button(window,text='x',width=10,height=5,command=multiply)
multiply.place(x=255,y=295)
minus=Button(window,text='-',width=10,height=5,command=minus)
minus.place(x=255,y=210)
plus=Button(window,text='+',width=10,height=5,command=plus)
plus.place(x=255,y=125)





window.mainloop()
```
